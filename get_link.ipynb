```python
#어떻게 생겼나 예시용
import json
with open('./bigData/crawling_netflix_1.json', 'r') as f:
    json_data = json.load(f)

df1=spark.createDataFrame(json_data['items'])
df_netflix=df1.drop('bloggername')
df_netflix=df_netflix.drop('bloggerlink')
df_netflix.toPandas()
```

```python
# crawling으로 저장한 파일 불러오기
# -> 제목(title 열)에 '추천'이라는 단어 있는 것(링크)만 저장하기
# 넷플릭스 관련 첫번째

import json
for a in range(1,12):
    with open('./bigData/crawling_netflix_'+str(a)+'.json', 'r') as f:
        json_data=json.load(f)
    df_netflix=spark.createDataFrame(json_data['items'])
    list_netflix_title=df_netflix.select(df_netflix['title']).collect()
    list_netflix_link=df_netflix.select(df_netflix['link']).collect()
    file_=open("./bigData/bloglink_netflix1.json","a+")
    for i in range(len(list_netflix_title)):
        for j in range(len(list_netflix_title[i][0].split())):
            if list_netflix_title[i][0].split()[j]==u'추천':
#                 print a,i
                file_.write(list_netflix_link[i][0])
                file_.write("\n")
    file_.close()
```
